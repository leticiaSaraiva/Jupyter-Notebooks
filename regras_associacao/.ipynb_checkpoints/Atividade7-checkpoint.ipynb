### Análise exploratória

```python
df.columns
```

```python
oc, qtd = np.unique(df.iloc[:,3], return_counts=True)
print(oc, qtd)
```

Mostrando a quantidade de ocorrências por cada tipo.

```python
for i, o in enumerate(oc):
    print(o, ': ', qtd[i])
```

Pegando as ocorrências que aconteceram pelo menos 4 vezes.

```python
qt = []
ind = []
for i, q in enumerate(qtd):
    if q > 5:
        qt.append(q)
        ind.append(i)
```

```python
oco = []
for i in ind:
    oco.append(oc[i])
```

Gráfico que mostra a quantidade de ocorrências para as ocorrências que aconteceram pelo menos 4 vezes.

```python
plt.figure(figsize=(18,6))

plt.xlabel('Tipo de ocorrência')
plt.ylabel('Quantidade de ocorrências')

plt.bar(oco, qt, color='r')
plt.show();
```

Plotando as ocorrências no espaço, usando latitude e longitude.

```python
import folium
```

```python
#Plotando marcadores indicando as ocorrências do dataframe
from folium.plugins import MarkerCluster

meio = [sum(df['LONGITUDE'])/len(df),sum(df['LATITUDE'])/len(df)+.02]

m = folium.Map(
    location=[meio[1],meio[0]],
    tiles='openstreetmap',
    zoom_start=11)

mc = MarkerCluster()
    
for index, c in df.iterrows():
    mc.add_child(folium.Marker([c['LATITUDE'], c['LONGITUDE']], 
              tooltip=c['NATUREZA DA OCORRÊNCIA'])).add_to(m)
m
```

```python
meio = [sum(df['LONGITUDE'])/len(df),sum(df['LATITUDE'])/len(df)+.02]

m = folium.Map(
    location=[meio[1],meio[0]],
    tiles='openstreetmap',
    zoom_start=11)

for i in range(0,len(df)):
    folium.Circle([df.iloc[i]['LATITUDE'], df.iloc[i]['LONGITUDE']], radius=10, 
                      color='red').add_to(m)
m
```

```python
#Plotando um mapa de calor (Heat Map)
from folium.plugins import HeatMap

lat = df.LATITUDE.tolist()
lng = df.LONGITUDE.tolist()

m = folium.Map(
    location=[meio[1],meio[0]],
    tiles='openstreetmap',
    zoom_start=11)

HeatMap(list(zip(lat, lng))).add_to(m)

m
```

Plotando a distribução das distâncias entre os pontos.

```python
from sklearn.metrics.pairwise import haversine_distances
import seaborn as sbn

localizacao = list(zip(df['LATITUDE'], df['LONGITUDE']))
raio_terra = 6371000/1000
distancias = haversine_distances(localizacao) * raio_terra
```

```python
dist = sorted(np.concatenate(distancias)) +[0]*len(localizacao)
i_loc = list(pd.Series(list(range(len(dist))))%2 == 0)

dist = pd.Series(dist)[i_loc]

sbn.distplot(dist);
```

### Algoritmos de clusterização

#### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(metric='precomputed', eps=40, min_samples=10).fit(distancias)
dbscan
```

```python
dbscan.labels_
```

Variando os parâmetros.

```python
dbscan1 = DBSCAN(metric='precomputed', eps=30, min_samples=6).fit(distancias)
dbscan1.labels_
```

```python
dbscan2 = DBSCAN(metric='precomputed', eps=60, min_samples=7).fit(distancias)
dbscan2.labels_
```

#### Clustering hierárquico

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clustering = AgglomerativeClustering(n_clusters=10, linkage='ward').fit(localizacao)
clustering 
```

```python
clustering.labels_
```

Variando os parâmetros.

```python
clustering1 = AgglomerativeClustering(n_clusters=8, linkage='ward').fit(localizacao)
clustering1.labels_
```

```python
clustering2 = AgglomerativeClustering(n_clusters=30, linkage='ward').fit(localizacao)
clustering2.labels_
```
